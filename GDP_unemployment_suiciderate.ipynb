```
import numpy as np
import matplotlib.pyplot as plt
```

```
GDP=500000000
N=120000000
n=10
gdp=np.zeros((n+1,1))
r=np.zeros((n+1,1))
u=np.zeros((n+1,1))

for i in range(n+1):
  a=(i/n)
  t=365
  gdp[i]=(GDP/365)*(a)*t
  r[i]=(gdp[i]-GDP)/GDP
  u[i]=(-0.1078*(r[i]))-0.0803
  print(i/n,' ',r[i],'  ',u[i])
```

```
plt.plot(r,u)
plt.xlabel('GDP growth rate')
plt.ylabel('Unemployment rate')
```

```
y=np.zeros((n+1,1))
for i in range(n+1):
  y[i]=(1949*u[i])+14345
  print(i/n,u[i],y[i])
plt.plot(u,y)

plt.xlabel('Unemployment rate')
plt.ylabel('Suicide rate')
```
